### In this section the list Game_table has all the info about each match:

````python
Game_table = []
Teams = []
for eachround in matches:
    Games = [eachround['sport_event']['competitors'][0]['abbreviation'], 
             eachround['sport_event_status']['home_score'],
             eachround['sport_event']['competitors'][1]['abbreviation'],
             eachround['sport_event_status']['away_score'], 
             [eachround['sport_event_status']['period_scores'][0]['home_score'], 
              eachround['sport_event_status']['period_scores'][0]['away_score']], 
             [eachround['sport_event_status']['period_scores'][1]['home_score'], 
              eachround['sport_event_status']['period_scores'][1]['away_score']]]
    Game_table.append(Games)
for game in Game_table:
    Teams.append(game[0])
    Teams.append(game[2])
pprint(Game_table[:10])
````

__B3.__ (5 points) Now, using the stored data for each match, construct aggregate result data for each team and store this in a list.

````python
uniqueTeams = list(set(Teams)) #we have 20 unique teams
#pprint(uniqueTeams)
GoalFor = []
for row in Game_table:
    GoalFor.append([row[0], row[1]])
    GoalFor.append([row[2], row[3]])
GoalAgainst = []
for row in Game_table:
    GoalAgainst.append([row[0], row[3]])
    GoalAgainst.append([row[2], row[1]])
````

````python
win = []
lose = []
draw = []
#in this part we are deciding on  the winners and losers:
for each in Game_table:
    if each[4][0]>each[4][1]:
        win.append(each[0])
        lose.append(each[2])
    elif each[4][0]==each[4][1]:
        draw.append(each[0])
        draw.append(each[2])
    else:
        lose.append(each[0])
        win.append(each[2]) 
for each in Game_table:
    if each[5][0]>each[5][1]:
        win.append(each[0])
        lose.append(each[2])
    elif each[5][0]==each[5][1]:
        draw.append(each[0])
        draw.append(each[2])
    else:
        lose.append(each[0])
        win.append(each[2])
#winners and losers lists + drawn's list:
winnerTeams = list(set(win))
loserTeams = list(set(lose))
drawTeams = list(set(draw))
````

````python
team_table = {}
for row in Game_table:
    Team = row[0]
    team_table.setdefault(Team, {})
    W = len([i for i in win if i == Team])
    D = len([i for i in draw if i == Team])
    L = len([i for i in lose if i == Team])
    G = W+D+L
    GF = sum([i[1] for i in GoalFor if i[0] == Team])
    GA = sum([i[1] for i in GoalAgainst if i[0] == Team])
    GD = GF-GA
    P = W*3+D*1
    team_table[Team]= [G, W, D, L, GF, GA, GD, P]
Soccer_List1 = [['Team', 'G', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'P']]
Soccer_List = []
for key in team_table.keys():
    info = team_table[key] 
    Soccer_List.append(
        [key, info[0], info[1], info[2], info[3], info[4], info[5], info[6], info[7]])
pprint(Soccer_List)
````
